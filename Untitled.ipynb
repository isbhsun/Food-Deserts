```python
import pandas as pd
import numpy as np
import zipfile
import matplotlib.pyplot as plt

from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split, KFold
from sklearn.linear_model import LinearRegression, Ridge, Lasso, LogisticRegression
from sklearn.neighbors import  KNeighborsRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, confusion_matrix, precision_score, recall_score, roc_curve, auc, roc_auc_score, f1_score
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV
# plt.style.use('ggplot')
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('data/food_insecurity.csv')
df.set_index('caseid2019', inplace=True)
```

```python
df_analysis = df.copy()
df_analysis.dropna(inplace=True)
df_analysis.info()
```

```python
y_SNAP = df_analysis[df_analysis['Applied for SNAP'] == 1]
n_SNAP = df_analysis[df_analysis['Applied for SNAP'] == 0]
```

```python
df_analysis
```

```python
y = df_analysis.pop('Applied for SNAP')
y_alt = df_analysis.pop('Applied SNAP but not received')
X = df_analysis
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, stratify=y, random_state=1)
```

```python
y_test.mean()
```

```python
snap_mean = y.mean()
want_mean = y_alt.mean()
print(want_mean)
print(snap_mean)
```

```python
rf = DecisionTreeClassifier(random_state=0)
rf.fit(X_train, y_train)

print("\n score:", rf.score(X_test, y_test))
y_predict = rf.predict(X_test)
print("\n confusion matrix:")
print(confusion_matrix(y_test, y_predict))
print("\n precision:", precision_score(y_test, y_predict))
print("    recall:", recall_score(y_test, y_predict))
print("f1:", f1_score(y_test, y_predict))
print("roc:", roc_auc_score(y_test, y_predict))

# precision = tp/(tp+fp)
# recall = tp/(tp+fn)
#feature importance
n = 10 # top 10 features

#importances = forest_fit.feature_importances_[:n]
importances = rf.feature_importances_
indices = np.argsort(importances)[::-1]
features = list(df_analysis.columns[indices])

# Print the feature ranking
print("\n Feature ranking:")

for f in range(n):
    print("%d. %s (%f)" % (f + 1, features[f], importances[indices[f]]))
```

```python
cross_val(rf, X_train, y_train, 'f1', 5)
```

```python
rf = RandomForestClassifier(random_state=0)
rf.fit(X_train, y_train)

print("\n score:", rf.score(X_test, y_test))
y_predict = rf.predict(X_test)
print("\n confusion matrix:")
print(confusion_matrix(y_test, y_predict))
print("\n precision:", precision_score(y_test, y_predict))
print("    recall:", recall_score(y_test, y_predict))
print("f1:", f1_score(y_test, y_predict))
print("roc:", roc_auc_score(y_test, y_predict))

# precision = tp/(tp+fp)
# recall = tp/(tp+fn)
#feature importance
n = 10 # top 10 features

#importances = forest_fit.feature_importances_[:n]
importances = rf.feature_importances_
std = np.std([rf.feature_importances_ for tree in rf.estimators_],
             axis=0)
indices = np.argsort(importances)[::-1]
features = list(df_analysis.columns[indices])

# Print the feature ranking
print("\n Feature ranking:")

for f in range(n):
    print("%d. %s (%f)" % (f + 1, features[f], importances[indices[f]]))
```

```python
# Plot the feature importances of the forest
fig, ax = plt.subplots(figsize=(12, 8))

ax.bar(range(n), importances[indices][:n], color="r", align="center")
ax.set_xticks(range(n))
ax.set_xticklabels(features[:n], rotation = 90)
ax.set_xlim([-1, n])
ax.set_xlabel("Importance")
ax.set_title("Feature Importances")
```
